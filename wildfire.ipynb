```python
import tensorflow as tf
import os
import cv2
import imghdr
from matplotlib import pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
from tensorflow.keras.layers import concatenate, Input
from tensorflow.keras.models import Model
```

```python
# Define the image input
image_input = Input(shape=(256, 256, 3), name='image_input')
conv1 = Conv2D(32, (3, 3), activation='relu')(image_input)
conv2 = Conv2D(64, (3, 3), activation='relu')(conv1)
flatten_img = Flatten()(conv2)
```

```python
# Define the numerical input
numerical_input = Input(shape=(2,), name='numerical_input')
dense1 = Dense(64, activation='relu')(numerical_input)
```

```python
concatenated = concatenate([flatten_img, dense1])
```

```python
dense_common = Dense(128, activation='relu')(concatenated)
output = Dense(2, activation='softmax')(dense_common)
```

```python
model = Model(inputs=[image_input, numerical_input], outputs=output)
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
